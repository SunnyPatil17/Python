### Pandas Vs Numpy
###### Numpy performs better for 50K rows or less. Pandas performs better than Numpy for 500K rows or more
###### Elements in Numpy arrays are accessed by their default integer position. Pandas Series Object is more flexible as you can define your own labeled index to index and access elements of an array.

```python
import pandas as pd
```

###### What kind of data does suit Pandas the most ? Ans - Tabular data, Arbitrary Matrix, Time Series data

###### Data-Set in Pandas. ------- For one Dimensional we use Series Object. For two Dimensional we use DataFrame. For three Dimensional we use Panel Data.

###### What is Series Object ? Ans - One Dimensional labeled array that contains data of similar or mixed types. 

```python
data = [1,2,3,4,5] 
srs1 = pd.Series(data)
print(srs1)
```

```python
# Passing 2D array into Series
data = [[1,2,3,4,5], [3,4,5]] 
srs2 = pd.Series(data)
print(srs2) # 2D array converted to 1D data
```

```python
print(type(srs1))
print(type(srs2))
```

```python
# Indexing the Panda Series. Customized Indexing. 
data = [1,2,3,4,5] 
srs3 = pd.Series(data, index=['a', 'b', 'c', 'd', 'e'])
print(srs3)
```

###### What is dataframe ? Ans - Two Dimensional labeled data structures with columns of potentially different types
###### Features of Dataframe ----- 1. Different column types 2. Mutable Size 3. Labeled Axes 4. Arithmetic Operations on rows and columns

```python
data = [1,2,3,4,5] 
df1 = pd.DataFrame(data)
df1
```

```python
data = [[1,2,3,4,5],[3,4,5]] 
df2 = pd.DataFrame(data)
df2
```

```python
data = {"a":[1,2,3,4,5],"b":[2,3,4,5,6] } 
df2 = pd.DataFrame(data)
df2
```

```python
data = {"Fruit":["apple","mango","banana"], "Count":[20,30,40]}
df3 = pd.DataFrame(data)
df3
```

```python
data = {"Fruit":["Orange"],"Count":30, "Price": 150}
df4 = pd.DataFrame(data)
df4
```

```python
pd.concat([df3,df4])
```

```python
pd.concat([df3,df4], axis = 0, join = "inner") # Common Column in output
```

```python
pd.concat([df3,df4], axis = 1, join = "inner") # Common Row in output
```

```python
pd.merge(df3,df4, how="outer")
```

```python
pd.merge(df3,df4, how="inner")
```

```python
# rename column
cars = cars.rename(columns = {'model':'model name'})
cars.head(2)
```

```python
cars.qsec = cars.qsec.fillna(cars.qsec.mean()) 
cars.head() # here initially we found that there is no null value and hence no change in output. 
# code to drop unwanted columns : cars = cars.drop(columns=['column header'])
# code to change mpg from string to float : cars.mpg = cars.mpg.astype(float)
```

```python
corr = cars[['mpg','cyl','disp','hp']].corr() # to find correlation between columns
corr
```

```python
# double up records in 'am' using lambda function
f = lambda x : x*2
cars['2am'] = cars['am'].apply(f)
cars.head()
```

```python
cars.sort_values(by='mpg').head(10) # sort the values of mpg column in ascending order
```

```python
cars.sort_values(by='mpg',ascending = False).head(10) # sort the values of mpg column in descending order
```

```python
filter_ = cars['cyl'] > 6 # filter records with more than 6 cylinders
filtered_cars = cars[filter_] # apply filter to dataframe
filtered_cars.head()
```
